```python
 # Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
file = "https://data.lacity.org/resource/d5tf-ez2w.csv"
```

```python
original_df = pd.read_csv(file)
original_df.head()
```

```python
original_df.info()
```

```python
renamed_df = original_df.rename(columns={"date_rptd":"date_reported", "date_occ":"date_occurred", "time_occ":"time_occurred", "rpt_dist_no":"reporting_district", "location":"address", "location_1":"location"})
renamed_df.head() 
```

```python
occurred_df = renamed_df.loc[renamed_df['date_occurred'].between('2019-01-01','2019-12-31', inclusive=True)]
sorted_df = occurred_df.sort_values("date_occurred", ascending=True)
sorted_df
```

```python
sorted_df.head()
```

```python
# victim age
plt.hist(sorted_df['vict_age'])
plt.xlabel('Age')
plt.ylabel('Counts')
plt.show()
print(sts.normaltest(sorted_df["vict_age"].sample(200)))
```

```python
# reporting district
plt.hist(sorted_df['reporting_district'])
plt.xlabel('District')
plt.ylabel('Counts')
plt.show()
print(sts.normaltest(sorted_df["reporting_district"].sample(100)))
```

```python
# time occurred
plt.hist(sorted_df['time_occurred'])
plt.xlabel('Time Occurred')
plt.ylabel('Counts')
plt.show()
print(sts.normaltest(sorted_df["time_occurred"].sample(100)))
```

```python
# date occurred
plt.hist(sorted_df["date_occurred"])
plt.xlabel('Date')
plt.ylabel('Counts')
plt.show()
```

```python
occurred_line = sorted_df.plot.line(x='date_occurred', y='vict_age')
```

```python
ax = sorted_df.plot.bar(x='vict_sex', y='vict_age',rot=0)
```
